```python
from ultralytics import YOLO

#Load a model
model = YOLO("yolo11n-cls.yaml").load("yolo11n-cls.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(data="/kaggle/input/newwww/data (1)", epochs=50, imgsz=512,close_mosaic = 0, cls = 10, dropout = 0.3, optimizer = 'SGD', lr0 = 0.0001)

# model = YOLO("runs/classify/train/weights/best.pt")

# Resume training
# results = model.train(resume=True)

metrics = model.val(data = "/kaggle/input/newwww/vald/vald")  # no arguments needed, dataset and settings remembered
metrics.top1  # top1 accuracy
metrics.top5  # top5 accuracy
```
